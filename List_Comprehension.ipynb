```python
#normal for loop
l = [1,2,3,4,5]
for i in l:
  print(i)

l2 = [i for i in l]
print(l2)
```

```python
#using list comprehension

l = [2, 4, 6, 8, 10, 114]
[i**2 for i in l if i**2>50]
```

```python
#given dictionary is consisted of vehicles and their
#construct a list of the names of vehicles with weights


dict = {"Sedan": 1500, "SUV": 2000, "Pickup": 2500}
d = [i.lower() for i in dict if dict[i]< 5000]
print(d)

```

get only the numbers in a sentence

```python
sentence = 'In 1984 there were 13 instances of a protest with over 100'
words = sentence.split()
result = [number for number in words if not number.isalpha()]
print(result)
```

find the common numbers in two lists (without using a tuple or set) list

```python
list_a = [1, 2, 3, 4]
list_b = [2, 3, 4, 5]
common = [a for a in list_a if a in list_b]
print(common)
```
